```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import statsmodels.api as sm
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# need to figure out where to bring in this CSV ... was previously in the same directory as this notebook
df_RPI_CPI = pd.read_csv('RPI_CPI.csv')
```

```python
df_RPI_CPI.set_index('Date', inplace=True)
df_RPI_CPI.index = pd.to_datetime(df_RPI_CPI.index)
```

```python
df_RPI_CPI.tail(12)
```

```python
df_RPI = df_RPI_CPI.drop('CPI', axis=1)
```

```python
df_RPI
```

```python
df_RPI.index = pd.to_datetime(df_RPI.index)
```

```python
df_RPI['Date'] = df_RPI.index
```

```python
df_RPI.index = pd.RangeIndex(start=0, stop=len(df_RPI))
```

```python
df_RPI
```

```python
dataframes = {}
est_frames = {}

for i in range(29):
    year = 88 + i
    dataframes[f'df_RPI_{year}_{year+5}'] = df_RPI[(year-87)*12:(year-87)*12+60]
    dataframes[f'df_RPI_{year+5}_{year+6}'] = df_RPI[(year-87)*12+60:(year-87)*12+72]
#     placeholder for est_frames here ... will be overwritten once we have estimates
    est_frames[f'df_RPI_{year+5}_{year+6}'] = dataframes[f'df_RPI_{year+5}_{year+6}']
                         
```

```python
mape_list = []
YOY_error_0_list = []

for i in range(20, 29):
    year = 88 + i
    df5y = dataframes[f'df_RPI_{year}_{year+5}']
    df5y1y = dataframes[f'df_RPI_{year+5}_{year+6}']
    
    model=sm.tsa.statespace.SARIMAX(df5y['RPI'],order=(4, 2, 1),seasonal_order=(1,1,1,12))
    results=model.fit()

    abc = results.predict(start=60,end=71,dynamic=True)
    aaa = pd.DataFrame(abc)
    est_frames[f'df_RPI_{year+5}_{year+6}'] = aaa.copy()

    fig1 = plt.plot(df5y1y['Date'],aaa['predicted_mean'])
    fig2 = plt.plot(df5y1y['Date'],df5y1y['RPI'])
    fig3 = plt.plot(df5y['Date'],df5y['RPI'])
    plt.show()

    mape = 0
    YOY_error_0 = 0
    for i in range(12):
        if i == 0:
            YOY_error_0 = (aaa[['predicted_mean']].iloc[i,0]/df5y.iloc[49,0]) - (df5y1y.iloc[i,0]/df5y.iloc[49,0]) 
        mape = mape + abs((aaa[['predicted_mean']].iloc[i,0]/df5y1y.iloc[i,0]) - 1)
        mape = mape/12
    mape_list.append(mape)
    YOY_error_0_list.append(YOY_error_0)
```

```python
bdbd = pd.DataFrame(mape_list)
print(f'average mape = {bdbd.mean()}')
bdbd.plot();
```

```python
dddd = pd.DataFrame(YOY_error_0_list)
print(f'average mape = {dddd.mean()}')
dddd.plot();
```

```python
threes = [(21, 22, 23),
         (24, 25, 26),
         (27, 28,29)]

def findall(element, matrix):
    result = []
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == element:
                result.append((i, j))
    return result


fig, axs = plt.subplots(3, 3, figsize=(16,10))

for i in range(20, 29):
    year = 88 + i

    df5y = dataframes[f'df_RPI_{year}_{year+5}']
    df5y1y = dataframes[f'df_RPI_{year+5}_{year+6}']
    es5y1y = est_frames[f'df_RPI_{year+5}_{year+6}']
    
    a_g = findall(i+1, threes)
    a_g[0][0]
    a_g[0][1]
    
    axs[a_g[0][0], a_g[0][1]].plot(df5y1y['Date'],es5y1y['predicted_mean'])
    axs[a_g[0][0], a_g[0][1]].plot(df5y1y['Date'],df5y1y['RPI'])
    axs[a_g[0][0], a_g[0][1]].plot(df5y['Date'],df5y['RPI']);
            
plt.savefig('SARIMAX-02_3x3.png')
```
